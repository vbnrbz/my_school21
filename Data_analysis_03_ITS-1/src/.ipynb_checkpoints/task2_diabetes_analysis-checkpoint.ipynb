## Задание 6. Подготовка данных для обучения моделей

### Замена NaN значений

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

transformed_data = imputer.fit_transform(df)
df = pd.DataFrame(transformed_data, columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome'])
```

Заполяем все NaN значения средними значениями, и обратно преобразуем в датафрейм из массива.

### Разделение данных на обучающую и тестовую выборки

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

В X переменную все столбцы, кроме Outcome. В Y только Outcome. Делим данные на выборки. test_size - на тесты будет выделено 30% данных, random_state=42 - фиксированное число для генератора случайных чисел, это обеспечивает одинаковый результат при каждом запуске программы. stratify=y - гарантирует, что классы в тренировочном и тестовом наборах распределены пропорционально, как в исходных данных.

### Проверка распределения целевого признака в обучающей и тестовой выборках

```python
print('Распределение признака в обучающей выборке:')
print(y_train.value_counts(normalize=True))

print('\nРаспределение признака в тестовой выборке:')
print(y_train.value_counts(normalize=True))
```

normalize=True - нормализует данные в долевом соотношении

### Масштабирование числовых признаков

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Масштабируем числовые признаки для того, что избежать доминирования одного признака над другим.

## Задание 7. Построение и оценка базовой модели логистической регрессии

### Обучение модели логистической регрессии на обучающей выборке

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)
```

### Создаем прогнозы на тестовой выборке

```python
y_pred = model.predict(X_test_scaled)
y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]
```

### Оцениваем точность модели

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-score: {f1:.2f}')
print(f'ROC-AUC: {roc_auc:.2f}')
```

Accuracy = 0.74: Модель правильно классифицирует 74% всех примеров, но эта метрика не всегда является наилучшей, если классы несбалансированы.
Precision = 0.66: Когда модель предсказывает наличие диабета, она ошибается в 34% случаев.
Recall = 0.53: Модель находит только 53% всех реальных случаев диабета. Это значит, что многие люди с диабетом остаются незамеченными.
F1-Score = 0.59: Это среднее гармоническое точности и полноты, и в данном случае оно говорит, что модель имеет компромиссное качество. Она не слишком хороша по обеим меткам.
ROC-AUC = 0.84: Модель в целом хорошо различает классы, с вероятностью 84% она правильно отличит положительный класс от отрицательного.

Модель имеет хорошую способность различать классы (ROC-AUC = 0.84), но нуждается в улучшении по меткам Recall и F1-score, чтобы повысить полноту и уменьшить количество пропущенных положительных случаев, что критично для задач, связанных с медицинскими диагнозами.

## Задание 8. Использование модели

### Вывод коэффициентов логистической регрессии

```python
coefficients = pd.DataFrame({'Feature': X_train.columns, 'Coefficient': model.coef_[0]}).sort_values(by='Coefficient', ascending=False)

print('Коэффициенты логистической регрессии:')
print(coefficients)
```

Признаки с положительными коэффициентами увеличивают вероятность диабета, с отрицательными вероятность уменьшается.

### Функция с прогнозом вероятности диабета

```python
def predict_diabetes(pregnancies, glucose, blood_pressure, skin_thickness, insulin, BMI, diabetes_pedigree_function, age):
    input_data = pd.DataFrame([[pregnancies, glucose, blood_pressure, skin_thickness, insulin, BMI, diabetes_pedigree_function, age]],
                         columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age'])
    scaled_data = scaler.transform(input_data)
    probability = model.predict_proba(scaled_data)[0, 1]
    prediction = model.predict(scaled_data)[0]

    return {'Probability': probability, 'Prediction': prediction}
```

```python
example = predict_diabetes(pregnancies=1, glucose=50, blood_pressure=50, skin_thickness=60, insulin=80, BMI=21.6, diabetes_pedigree_function=0.5, age=30)
print(f"Вероятность диабета: {example['Probability']:.4f}.")
if not example['Prediction']:
    prediction_result = "Нет диабета"
else:
    prediction_result = "Есть диабет"
print(f'По данным: {prediction_result}.')
```

## Бонусное задание 9. Подбор гиперпараметров с использованием Grid Search

### Определение сетки гиперпараметров для логистической регрессии и выбор метода регуляции

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

params = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'penalty': ['l1', 'l2'],
    'solver': ['liblinear']  # Для L1-регуляризации нужен solver='liblinear'
}

model = LogisticRegression(random_state=42)

grid_search = GridSearchCV(model, params, scoring='roc_auc', cv=5)
grid_search.fit(X_train_scaled, y_train)
print(f'Лучшие параметры: {grid_search.best_params_}')

y_pred = grid_search.predict(X_test_scaled)
auc_score = roc_auc_score(y_test, y_pred)
print(f'Лучший ROC-AUC: {auc_score:.4f}')
```

### Переобучаем модель

```python
best_model = grid_search.best_estimator_

model = LogisticRegression(random_state=42)
model.fit(X_train_scaled, y_train)
```

### Оценка производительности улучшенной модели

```python
best_train_pred = best_model.predict(X_train_scaled)
best_train_prob = best_model.predict_proba(X_train_scaled)[:, 1]
best_test_pred = best_model.predict(X_test_scaled)
best_test_prob = best_model.predict_proba(X_test_scaled)[:, 1]

print('Тестовая выборка:')

best_test_accuracy = accuracy_score(y_test, best_test_pred)
best_test_precision = precision_score(y_test, best_test_pred)
best_test_recall = recall_score(y_test, best_test_pred)
best_test_f1 = f1_score(y_test, best_test_pred)
best_test_roc_auc = roc_auc_score(y_test, best_test_prob)

print(f"Accuracy: {best_test_accuracy:.2f}")
print(f"Precision: {best_test_precision:.2f}")
print(f"Recall: {best_test_recall:.2f}")
print(f"F1-score: {best_test_f1:.2f}")
print(f"ROC-AUC: {best_test_roc_auc:.2f}")

print('\nОбучающая выборка:')

best_train_accuracy = accuracy_score(y_train, best_train_pred)
best_train_precision = precision_score(y_train, best_train_pred)
best_train_recall = recall_score(y_train, best_train_pred)
best_train_f1 = f1_score(y_train, best_train_pred)
best_train_roc_auc = roc_auc_score(y_train, best_train_prob)

print(f"Accuracy: {best_train_accuracy:.2f}")
print(f"Precision: {best_train_precision:.2f}")
print(f"Recall: {best_train_recall:.2f}")
print(f"F1-score: {best_train_f1:.2f}")
print(f"ROC-AUC: {best_train_roc_auc:.2f}")
```

### Сравнение с первоначальной моделью

```python
model_train_pred = model.predict(X_train_scaled)
model_train_prob = model.predict_proba(X_train_scaled)[:, 1]
model_test_pred = model.predict(X_test_scaled)
model_test_prob = model.predict_proba(X_test_scaled)[:, 1]

model_test_accuracy = accuracy_score(y_test, model_test_pred)
model_test_precision = precision_score(y_test, model_test_pred)
model_test_recall = recall_score(y_test, model_test_pred)
model_test_f1 = f1_score(y_test, model_test_pred)
model_test_roc_auc = roc_auc_score(y_test, model_test_prob)

model_train_accuracy = accuracy_score(y_train, model_train_pred)
model_train_precision = precision_score(y_train, model_train_pred)
model_train_recall = recall_score(y_train, model_train_pred)
model_train_f1 = f1_score(y_train, model_train_pred)
model_train_roc_auc = roc_auc_score(y_train, model_train_prob)

print('Улучшение модели (test):')

delta_test_accuracy = best_test_accuracy - model_test_accuracy
delta_test_precision = best_test_precision - model_test_precision
delta_test_recall = best_test_recall - model_test_recall
delta_test_f1 = best_test_f1 - model_test_f1
delta_test_roc_auc = best_test_roc_auc - model_test_roc_auc

print('*'*50)
print(f"Accuracy: {delta_test_accuracy}")
print(f"Precision: {delta_test_precision}")
print(f"Recall: {delta_test_recall}")
print(f"F1-score: {delta_test_f1}")
print(f"ROC-AUC: {delta_test_roc_auc}")

print('\nУлучшение модели (train):')

delta_train_accuracy = best_train_accuracy - model_train_accuracy
delta_train_precision = best_train_precision - model_train_precision
delta_train_recall = best_train_recall - model_train_recall
delta_train_f1 = best_train_f1 - model_train_f1
delta_train_roc_auc = best_train_roc_auc - model_train_roc_auc

print('*'*50)
print(f"Accuracy: {delta_train_accuracy}")
print(f"Precision: {delta_train_precision}")
print(f"Recall: {delta_train_recall}")
print(f"F1-score: {delta_train_f1}")
print(f"ROC-AUC: {delta_train_roc_auc}")
```
